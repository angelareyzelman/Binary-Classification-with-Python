```python
import numpy as np
import matplotlib.pyplot as plt

x0_x = []
x0_y = []
x1_x = []
x1_y = []
for i in range(100):
    x0_x.append(0)
    x0_y.append(np.random.uniform(0,1))
    x1_x.append(5)
    x1_y.append(np.random.uniform(0,1))
    
fig1 = plt.figure(num=1, clear=True)
ax1 = fig1.add_subplot(1,1,1)
#scatter the original points
ax1.scatter(x0_x, x0_y, c="blue")
ax1.scatter(x1_x, x1_y, c="orange")

fig2 = plt.figure(num=2, clear=True)
ax2 = fig2.add_subplot(1,1,1)

ax2.scatter(x0_x[0:20], x0_y[0:20], c="blue", marker=".", s=40)
ax2.scatter(x1_x[0:20], x1_y[0:20], c="blue", marker=".", s=40)
ax2.scatter(x0_x[21:100], x0_y[21:100], c="orange", marker="+", s=20)
ax2.scatter(x1_x[21:100], x1_y[21:100], c="orange", marker="+", s=20)


def euclidian_distance(a,b, p=2):
    a=[]
    b=[]
    for k in range(len(b)):
        distance = 0
        distance = (abs(a[k] - b[k])**p)**(1/p)
        return distance

testpoints = []
testpoints += x0_x[0:20], x0_y[0:20], x1_x[0:20], x1_y[0:20]
print(testpoints) 

distances=[]
for i in x0_y[0:20]:
    distances.append(euclidian_distance(i,x0_y[21:100]))
for k in x1_y[0:20]:
    distances.append(euclidian_distance(k,x1_y[21:100]))
print(distances)

xcoords=[]
ycoords=[]  
fig4=plt.figure(num=4, clear=True)
ax4 = fig4.add_subplot(1,1,1)
for k in range(200):
    xcoords.append(np.random.uniform(0,1))
    ycoords.append(np.random.uniform(0,1))
    num = np.random.uniform(0,1)
    print(num)
    if num > 0.5:
        ax4.scatter(xcoords[k], ycoords[k], c="orange", marker='.')
    else:
        ax4.scatter(xcoords[k], ycoords[k], c="blue", marker='.')

# x_0=[]
# y_0=[]
# fig5 = plt.figure(num=5, clear=True)
# ax5 = fig5.add_subplot(1,1,1)
# ax5.scatter(x_0[0:20], y_0[0:20], c="red", marker="v", s=20)
# ax5.scatter(x_0[21:100], y_0[21:100], c="black", marker="*", s=20)
```
